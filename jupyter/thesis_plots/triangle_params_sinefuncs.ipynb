```python
from floquet_simulations.plot_functions import *
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from numpy import pi
from pathlib import Path
import numpy as np
import matplotlib as mpl
cm_unit = 1/2.54

figsize=(4*(2/3)*cm_unit,4*cm_unit)
fontsize = 10
PlotParams(fontsize=fontsize)
num_sites = 13
axes_tick_pos=np.array([0,2,4,6,8,10,12])
bbox_colourbar_anchor = (0.9, 0.57, 1, 0.73)
colourbar_pad=0.1
colourbar_size_percentage=5
xlabel_coords = [0.52,-0.11]
ylabel_coords = [-0.14,0.5]

save_fig_dir = Path().absolute().parent.parent/"figures"/"thesis"/"floquet"
```

```python
x = np.linspace(0,4*np.pi,100)
fig, ax = plt.subplots(figsize=figsize)
A2 = 1.7
A3 = 1
xtick_labels = ["$0$", "$T$", "$2T$"]
xticks = [0, 2*pi, 4*pi]
ax.axhline(y=0, color='0', lw=0.5)
ax.plot(x, A2*np.sin(x), label="$W_2(t)$", color="#C30934")
ax.plot(x, A3*np.sin(x), linestyle="--", label="$W_3(t)$", color="#0F1980")

ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, 
        labeltop=False, direction = "in", length = 3,  pad=2)  
ax.tick_params(axis="y", left=True, right=False, labelleft=True, 
        labelright=False, direction = "in", length = 3, pad=2)  

ax.set_yticks([0, A2, A3])
ax.set_yticklabels([r"$0$", r"$A_2$", r"$A_3$"])
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels)

leg = ax.legend(loc='upper right', bbox_to_anchor=(1.05,1.1))
ax.set_xlabel("$t$")
ax.xaxis.set_label_coords(1.06,0.53)
ax.set_xlim([0,4*pi])

fig.savefig(save_fig_dir/"triangle_parameters_beta=1_cosine.pdf", format="pdf", bbox_inches="tight", pad_inches=0.04)
plt.show()

fig, ax = plt.subplots(figsize=figsize)
A2 = 1.7
A3 = 1
ax.axhline(y=0, color='0', lw=0.5)
ax.plot(x, A2*np.sin(x), label="$W_2(t)$", color="#C30934")
ax.plot(x, A3*np.sin(2*x), linestyle="--",label="$W_3(t)$", color="#0F1980")

ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, 
        labeltop=False, direction = "in", length = 3,  pad=2)  
ax.tick_params(axis="y", left=True, right=False, labelleft=True, 
        labelright=False, direction = "in", length = 3, pad=2)  

ax.set_yticks([0, A2, A3])
ax.set_yticklabels([r"$0$", r"$A_2$", r"$A_3$"])
ax.set_xticks(xticks)
ax.set_xlabel("$t$")
ax.xaxis.set_label_coords(1.06,0.53)
ax.set_xlim([0,4*pi])

ax.set_xticklabels(xtick_labels)
leg = ax.legend(loc='upper right', bbox_to_anchor=(1.05,1.1))
fig.savefig(save_fig_dir/"triangle_parameters_beta=2_cosine.pdf", format="pdf", bbox_inches="tight", pad_inches=0.04)
plt.show()
```
