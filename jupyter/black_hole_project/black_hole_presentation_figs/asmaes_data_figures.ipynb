```python
import pandas as pd
from pathlib import Path
read_dir = Path().absolute().parent.parent.parent/"data"/"black_hole"/"asmae_data"
import numpy as np
import matplotlib.pyplot as plt
from floquet_simulations.plot_functions import PlotParams
fontsize=9
PlotParams(fontsize=9)
save_dir = Path().absolute().parent.parent.parent/"figures"/"black_hole_pres"
figsize = (4.4, 4.4)
cm_unit = 1/2.54
#for b
n_b = 6
energies0 = pd.read_csv(read_dir/'energies_a_new_0.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
energies1 = pd.read_csv(read_dir/'energies_a_new_1.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
energies2 = pd.read_csv(read_dir/'energies_a_new_2.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
energies3 = pd.read_csv(read_dir/'energies_a_new_3.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
ps0 = pd.read_csv(read_dir/'ps_a_new_0.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
ps1 = pd.read_csv(read_dir/'ps_a_new_1.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
ps2 = pd.read_csv(read_dir/'ps_a_new_2.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]
ps3 = pd.read_csv(read_dir/'ps_a_new_3.txt', sep=" ", header=None).to_numpy().flatten()[:n_b]

# for c
time_data0 = pd.read_csv(read_dir/'fig3_c_time_data_a=1.0.txt', sep=" ", header=None).to_numpy().flatten()
time_data1 = pd.read_csv(read_dir/'fig3_c_time_data_a=1.6666666666666665.txt', sep=" ", header=None).to_numpy().flatten()
time_data2 = pd.read_csv(read_dir/'fig3_c_time_data_a=2.333333333333333.txt', sep=" ", header=None).to_numpy().flatten()
time_data3 = pd.read_csv(read_dir/'fig3_c_time_data_a=3.0.txt', sep=" ", header=None).to_numpy().flatten()

pop_data0 = pd.read_csv(read_dir/'fig3_c_population_data_a=1.0.txt', sep=" ", header=None).to_numpy().flatten()
pop_data1 = pd.read_csv(read_dir/'fig3_c_population_data_a=1.6666666666666665.txt', sep=" ", header=None).to_numpy().flatten()
pop_data2 = pd.read_csv(read_dir/'fig3_c_population_data_a=2.333333333333333.txt', sep=" ", header=None).to_numpy().flatten()
pop_data3 = pd.read_csv(read_dir/'fig3_c_population_data_a=3.0.txt', sep=" ", header=None).to_numpy().flatten()

#for d
temp0 = pd.read_csv(read_dir/'fig3_d_temps_a=1.0.txt', sep=" ", header=None).to_numpy().flatten() 
temp1 = pd.read_csv(read_dir/'fig3_d_temps_a=1.6666666666666665.txt', sep=" ", header=None).to_numpy().flatten() 
temp2 = pd.read_csv(read_dir/'fig3_d_temps_a=2.333333333333333.txt', sep=" ", header=None).to_numpy().flatten() 
temp3 = pd.read_csv(read_dir/'fig3_d_temps_a=3.0.txt', sep=" ", header=None).to_numpy().flatten()
tfin0 = pd.read_csv(read_dir/'fig3_d_tfin_a=1.0.txt', sep=" ", header=None).to_numpy().flatten()
tfin1 = pd.read_csv(read_dir/'fig3_d_tfin_a=1.6666666666666665.txt', sep=" ", header=None).to_numpy().flatten()
tfin2 = pd.read_csv(read_dir/'fig3_d_tfin_a=2.333333333333333.txt', sep=" ", header=None).to_numpy().flatten()
tfin3 = pd.read_csv(read_dir/'fig3_d_tfin_a=3.0.txt', sep=" ", header=None).to_numpy().flatten()

TH0 = 1/4/np.pi
TH1 = 5/3/4/np.pi
TH2 = 7/3/4/np.pi
TH3 = 3/4/np.pi

colour0 = "#0f1980"#"#1565C0"
colour1 = "#006F63"
colour2 = "#F78320"
colour3 = "#A91401"
marker0 = "*"
marker1 =  "+"
marker2 = "o"
marker3 = "x"
markersize = 4
markersize_p1 = 2
```

```python
m0, b0 = np.polyfit(energies0[1:], np.log(ps0[1:]), 1)
m1, b1 = np.polyfit(energies1[1:], np.log(ps1[1:]), 1)
m2, b2 = np.polyfit(energies2[1:], np.log(ps2[1:]), 1)
m3, b3 = np.polyfit(energies3[1:], np.log(ps3[1:]), 1)
linewidth = 0.5
fig, ax = plt.subplots(figsize = (figsize[0]*cm_unit, figsize[1]*cm_unit))
ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, 
        labeltop=False, direction = "in", length = 3, pad=1)  
ax.tick_params(axis="y", left=True, right=False, labelleft=True, 
        labelright=False, direction = "in", length = 3, pad=1)  
ax.semilogy( energies0, ps0, 'o', markersize=markersize_p1, color=colour0)
ax.semilogy( energies1, ps1, 'o', markersize=markersize_p1, color=colour1)
ax.semilogy( energies2, ps2, 'o', markersize=markersize_p1, color=colour2)
ax.semilogy( energies3, ps3, 'o', markersize=markersize_p1, color=colour3)
ax.plot(energies0, np.exp(m0*energies0 + b0), color=colour0, linestyle='-', linewidth=linewidth, label=rf"$\tilde{{T}}_H = {-1/m0:.3f}$")
ax.plot(energies1, np.exp(m1*energies1 + b1), color=colour1, linestyle='-', linewidth=linewidth,label=rf"$\tilde{{T}}_H = {-1/m1:.3f}$")
ax.plot(energies2, np.exp(m2*energies2 + b2), color=colour2, linestyle='-', linewidth=linewidth,label=rf"$\tilde{{T}}_H = {-1/m2:.3f}$")
ax.plot(energies3, np.exp(m3*energies3 + b3), color=colour3, linestyle='-', linewidth=linewidth,label=rf"$\tilde{{T}}_H = {-1/m3:.3f}$")
ax.set_xlim([0,5.4])
ax.set_xticks([0,1,2,3,4,5])
ax.set_ylabel(r"$P_n$", rotation=0)
ax.set_xlabel(r"$E_n$")
plt.legend(fontsize=6, bbox_to_anchor=(0,0.03,1,1))
ax.yaxis.set_label_coords(-0.12,0.92)
ax.xaxis.set_label_coords(0.5,-0.08)
# fig.savefig(save_dir/"probabilities.pdf", format="pdf", bbox_inches="tight", pad_inches=0.03)
fig.savefig(save_dir/"probabilities.png", format="png", bbox_inches="tight", pad_inches=0.03,dpi=600)
plt.show()
```
