```python
#bessel
num_sites = 5
centres = [2]
funcs = [Cosine]
a0 = 40
phi0 = 0
onsite0 =0
num_omegas = 200
hoppings = np.empty(num_omegas, dtype=np.complex128)

omega0s = np.linspace(4,40,num_omegas)
for i, omega0 in enumerate(omega0s):
    T = 2*pi/omega0
    params = [[a0, omega0, phi0, onsite0]]
    _, HF = CreateHFGeneral(num_sites, centres, funcs, params, T, circle_boundary=0)
    hoppings[i] = HF[2,3]
```

```python
from scipy.special import jv
cm_unit = 1/2.54
plt.rc('text.latex', preamble=r'\usepackage{amsmath,nicefrac,xfrac}')
figsize_rec = (3.9, 3.9/2)
fig, ax = plt.subplots(figsize=(figsize_rec[0]*cm_unit, figsize_rec[1]*cm_unit))
ax.plot(1/omega0s, -np.real(hoppings), color='#006F63')
ax.plot(1/omega0s, jv(0, 40/omega0s), ':', color="#C30934", label=r"$\mathcal{J}_0(40/\omega)$")
ax.set_xlabel(r"$\nicefrac{1}{\omega}$", fontsize=14, labelpad=0.1)
ax.set_ylabel(r"$\nicefrac{{H_S^{t_0}}_{j,j+1}}{-J}$", fontsize=14)#
# plt.legend()
bessel_plot_dir =  Path().absolute().parent.parent/"figures"/"paper"/"bessel_func.pdf"
fig.savefig(bessel_plot_dir, format="pdf", bbox_inches="tight")
plt.show()
```
