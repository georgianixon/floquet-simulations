```python
#bessel
num_sites = 5
centres = [2]
funcs = [Cosine]
a0 = 40
phi0 = 0
onsite0 =0
num_omegas = 200
hoppings = np.empty(num_omegas, dtype=np.complex128)

omega0s = np.linspace(4,40,num_omegas)
for i, omega0 in enumerate(omega0s):
    T = 2*pi/omega0
    params = [[a0, omega0, phi0, onsite0]]
    _, HF = CreateHFGeneral(num_sites, centres, funcs, params, T, circle_boundary=0)
    hoppings[i] = HF[2,3]
```
