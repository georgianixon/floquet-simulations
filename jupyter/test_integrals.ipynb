```python
import numpy as np
import scipy.integrate as integrate
from math import pi
from numpy import cos, sin, sqrt
from scipy.special import struve, gamma, jv
import matplotlib.pyplot as plt

```

```python
f = -0.4
phi_offset = pi/12
omega = 8
results = []
bessel_res = []
for phi_offset in [0, pi/12, pi/3, pi, 13*pi/7, 19*pi/7]:
    for f in [-10, -1, -0.5, 0.5, 1, 10]:
        results.append((omega*(integrate.quad(lambda t:
                                         cos(f*cos(t*omega+phi_offset)), 0, 2*pi/omega)[0])))
        bessel_res.append(((2*pi)*jv(0,f)))
# struve_res = struve(0,f)
print(results)
print(bessel_res)
# print("integral result", result.round(5))
# "struve func of f", struve_res)[]
```
